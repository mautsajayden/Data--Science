Housing Prices 

```python
import pandas as pd 

#load data 

housing_file_path = "melb_data.csv"

housing_data = pd.read_csv(housing_file_path)

#filtering the data and replacing missing values 

filtered_housing_data = housing_data.dropna(axis=0)

housing_data = filtered_housing_data
```

The prediction target for the housing prices

```python
y = housing_data.Price

feature_columns = ['LotArea','YearBuilt','1stFlrSF','2ndFlrSF','FullBath','BedroomAbvGr','TotRmsAbvGrd']

X = housing_data[feature_columns]
```

Making the model for the housing

```python
from sklearn.tree import DecisionTreeRegressor as d_t_r

housing_model = d_t_r(random_state=1)

housing_model.fit(X,y)

print("First in-sample predictions: ", housing_model.predict(X.head()))
print("Actual target valkues for those homes: ", y.head().tolist)
```

Splitting the data                                                                                                                                                                              
Housing features are in X and the target is y

```python
from sklearn.model_selection import train_test_split as TTS 

train_X,val_X, train_y,val_y = TTS(X,y,random_state=1) 

#making the model 
housing_model = d_t_r(random_state=1)

#fitting the model 
housing_model.fit(train_X,train_y)
```

Making predictions with the validation data 

```python
val_predictions = housing_model.predict(val_X) 
```

Caculating the Mean Abosolute Error in Validation Data

```python
from sklearn.metrics import mean_absolute_error as MAE

val_error = MAE(val_y,val_predictions)

print(val_error)
```
